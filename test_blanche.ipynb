```python
import pandas as pd
```

```python
df = pd.read_csv("data/genres_count_over_time.csv", names=["year", "genre", "count"], header=0)
```

```python
df.head()
```

```python
df["year"] = df.apply(lambda x: int(x.year), axis=1)
```

```python
df_pivoted = df.pivot(index="genre", columns="year", values="count")
```

```python
df_pivoted.fillna(0, inplace=True)
```

```python
df_pivoted.head()
```

```python
df_pivoted.apply(lambda x: x.apply(lambda y: int(y)), axis=1)
```
